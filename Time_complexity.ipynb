```python
import matplotlib.pyplot as plt
import pandas as pd 
import time
import numpy as np
```

```python
log = pd.read_csv('new_time_complexity_python.csv')
```

```python
log
```

```python
sub_array = log.drop(['Divide_Conquer_time2','Strassen_optimized_2'],axis=1)
```

```python
sub_array.to_csv('sub.csv')
```

```python
ax = sub_array.plot(x='Array_size',title='Time Complexity Comparison',figsize=(10,10),fontsize=15)
ax.set_xlabel("Array Size",fontsize=15)
ax.set_ylabel("Time Duration",fontsize= 15)
ax.legend(fontsize=15)
plt.savefig('time_complexity.png',dpi=300)
```

```python
sub_strassen = log.drop(['Schoolbook_time', 'Divide_Conquer_time',
       'Divide_Conquer_time2'],axis=1)
```

```python
sub_strassen.to_csv("sub_strassen.csv")
```

```python
ax = sub_strassen.plot(x='Array_size',title='Time Complexity Comparison between Strassen',figsize=(10,10),fontsize=15)
ax.set_xlabel("Array Size",fontsize=15)
ax.set_ylabel("Time Duration",fontsize= 15)
ax.legend(fontsize=15)
plt.savefig('time_complexity_strassen.png',dpi=300)
```
